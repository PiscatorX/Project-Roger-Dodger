```r
library(tidyverse)
library(caret)
library(broom)
library(MASS)
```

```r
#import data
#Zeolite April database
zeolite_data_april <- read.table("C:/Users/DrewX/Documents/Project-Roger-Dodger/Python-ML/zeolites database catagories V 2_imputed.txt", sep ="\t", header = T )
#Zeolite June database
zeolite_data_june <- read.table("C:/Users/DrewX/Documents/Project-Roger-Dodger/Python-ML/zeolites database catagories june data base_imputed.txt", sep ="\t", header = T)
```

```r
#train our scaler 
Scaler_april <- preProcess(zeolite_data_april, method = c("center", "scale"))
Scaler_june <- preProcess(zeolite_data_june, method = c("center", "scale"))
```

```r
#generatue an dataframe of scaled variables
zeolite_april <- predict(Scaler_april, zeolite_data_april)
zeolite_june <- predict(Scaler_june, zeolite_data_june)
```

```r
fit_april <- lm(Capacity ~ ., data = zeolite_april)
(fit_april_summary <- summary(fit_april))
```

```r
fit_june <- lm(Capacity ~ ., data = zeolite_june) 
(fit_june_summary <- summary(fit_june)) 
```

```r
stepwise_april <- stepAIC(fit_april , direction = "both")
```

```r
summary(stepwise_april)
```

```r
stepwise_june <- stepAIC(fit_june  , direction = "both")
```

```r
summary(stepwise_june)
```

```r
june_summary <- summary(stepwise_june) %>% tidy() %>% arrange(desc(estimate))
```

```r
#################################################################################
#  This model used data where the categorical variables have been encoded from Python 
#   
#
#################################################################################
June_data_encoded <- read.table("C:/Users/DrewX/Documents/Project-Roger-Dodger/Python-ML/ZeoX_Final_encoded_V2x.tsv", sep ="\t", header = T )
ref_numeric <- zeolite_data_june %>% select_if(is.numeric) %>% colnames()
Scaler_june_encoded <- preProcess(June_data_encoded, method = list(center = ref_numeric, scale = ref_numeric))
June_encoded <- predict(Scaler_june_encoded, June_data_encoded)
```

```r
colnames(June_data_encoded)
```

```r
fit_encoded <- lm(Capacity ~ ., data = June_encoded)
(fit_encoded_summary <- summary(fit_encoded))
```

```r
stepwise_encoded <- stepAIC(fit_encoded, direction = "both")
```

```r
summary(stepwise_encoded)
```

```r
encoded_summary <- summary(stepwise_encoded) %>% tidy() %>% arrange(desc(estimate))
```
